# Remove the unnecessary columns from our dataset. The following columns are not needed in our tasks but are also empty since the data was removed for privacy reasons.

We now want to unpivot our dataframe on the question columns. We will do so by using the melt function.

```python
 # The id section will cover the demographic information.
id_vars_data = manipulated_data.columns[0:8]
id_vars_data
```

```python
value_vars_data = manipulated_data.columns[8:]
value_vars_data
```

```python
melted_data = pd.melt( manipulated_data , id_vars = id_vars_data , value_vars = value_vars_data,  var_name = "Questions", value_name = "Response"  )
melted_data
```

# Tasks, the client is inquiring about.

- How many people responded per question?
- what was the original question?
- How many people provided the same response?

```python
# We need to create a column to identify the question.

questions_data = pd.read_excel(  pwd +  "/data.xlsx" , sheet_name = 'Part_one'   )
questions_data
```

```python
questions_data.drop(  columns = [ "QUESTIONS" , "SUB-QUESTIONS" , "SUB_QUESTIONS"  ] , inplace = True   )
questions_data
```

```python
# merge this dataframe to our melted_data dataframe to obtain the Question column in our main dataframe.

merged_dataset = pd.merge( melted_data ,questions_data , how = "inner" ,left_on = "Questions" , right_on  = "QUESTIONS + SUB_QUESTIONS" )
merged_dataset
```

```python
# cleaning the merged_dataset.

merged_dataset.drop( columns= "Questions" , inplace = True )
```

```python
merged_dataset
```

```python
merged_dataset.rename( columns =  {"QUESTIONS.1":"Questions" , "QUESTIONS + SUB_QUESTIONS":"Questions + Response"}, inplace = True  )
```

```python
merged_dataset

# Created a column to show the original question. 
```

# Number of response per question.

```python

response = merged_dataset[merged_dataset["Response"].notnull() ].groupby("Questions")["Respondent ID"].nunique().to_frame("total_response").reset_index()
            # Gets data where response is not null/empty
```

```python
response
```

```python
agg_dataset = pd.merge(  merged_dataset , response , how = "inner" , on = "Questions"     )
agg_dataset
```

```python
agg_dataset["Respondent ID"].nunique()
```

# How many people provided the same response per question.


```python
same_response = agg_dataset.groupby(["Questions + Response","Response"])["Respondent ID"].nunique().to_frame("Num_of_Same_Response").reset_index()
same_response
```

```python
final_dataset = pd.merge(agg_dataset , same_response , how = 'left' , left_on = ["Questions + Response", "Response"] , right_on = ["Questions + Response", "Response"]       )
```

```python
final_dataset
```

```python
final_dataset["Num_of_Same_Response"].fillna(0 , inplace = True)
# fill the empty values of the columns num_of_same_response with 0
final_dataset
```

# Clean the dataset, by renaming the columns

We can also look at gender. 

```python
gender = output.groupby(['Division Primary','Gender'])['Respondent ID'].nunique().to_frame('gender_response').reset_index()
gender
```

```python
gender['total'] = gender.groupby('Division Primary')['gender_response'].transform('sum')
# We could have also merged this gender dataframe with the Division dataframe to obtain the same total column values.
```

```python
gender['rank'] = gender['total'].rank(method = 'dense' , ascending = False)
# Creates a rank based on the amount of people in the division.
```

```python
gender.dtypes
```

```python
gender['rank'] = gender['rank'].astype(int)
```

```python
gender.sort_values(by = 'rank', ascending=False)
```

We see which division had the biggest population by the rank column, and see the total respondents per division by the total column. Data shows the different genders that are present in the primary divisions. Data may be useful to people who are applying for jobs and want to see the diverse genders in the division.
